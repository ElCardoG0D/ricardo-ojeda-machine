```python
import pandas as pd
from pathlib import Path
from kedro.framework.startup import bootstrap_project
from kedro.framework.session import KedroSession

pd.set_option("display.max_columns", None)

# üöÄ Ruta ra√≠z del proyecto
project_path = Path("C:/Users/Ricardo/ricardo-ojeda-machine")

# Bootstrap para inicializar proyecto
bootstrap_project(project_path)

# üöÄ Crear sesi√≥n forzando conf_source
with KedroSession.create(
    project_path=project_path,
    conf_source=str(project_path / "conf")
) as session:
    context = session.load_context()
    catalog = context.catalog

    # Cargar datasets
    intakes = catalog.load("intakes")
    outcomes = catalog.load("outcomes")
    licenses = catalog.load("licenses")

print("üìÇ Intakes:", intakes.shape)
print("üìÇ Outcomes:", outcomes.shape)
print("üìÇ Licenses:", licenses.shape)

```

```python
# ----------------------------
# PRIMERAS FILAS
# ----------------------------
print("\nüîé Intakes:")
print(intakes.head())

print("\nüîé Outcomes:")
print(outcomes.head())

print("\nüîé Licenses:")
print(licenses.head())

# ----------------------------
# INFO GENERAL
# ----------------------------
print("\n‚ÑπÔ∏è Info Intakes")
intakes.info()

print("\n‚ÑπÔ∏è Info Outcomes")
outcomes.info()

print("\n‚ÑπÔ∏è Info Licenses")
licenses.info()

# ----------------------------
# VALORES NULOS
# ----------------------------
print("\n‚ùå Nulos Intakes:\n", intakes.isnull().sum())
print("\n‚ùå Nulos Outcomes:\n", outcomes.isnull().sum())
print("\n‚ùå Nulos Licenses:\n", licenses.isnull().sum())

# ----------------------------
# ESTAD√çSTICAS DESCRIPTIVAS
# ----------------------------
print("\nüìä Estad√≠sticas descriptivas - Intakes")
print(intakes.describe(include="all"))

print("\nüìä Estad√≠sticas descriptivas - Outcomes")
print(outcomes.describe(include="all"))

print("\nüìä Estad√≠sticas descriptivas - Licenses")
print(licenses.describe(include="all"))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Boxplot edad al ingreso
if "Age upon Intake" in intakes.columns:
    plt.figure(figsize=(8,4))
    sns.boxplot(x="Age upon Intake", data=intakes)
    plt.title("Boxplot - Edad al ingreso")
    plt.show()

# Boxplot edad al outcome
if "Age upon Outcome" in outcomes.columns:
    plt.figure(figsize=(8,4))
    sns.boxplot(x="Age upon Outcome", data=outcomes)
    plt.title("Boxplot - Edad al egreso")
    plt.show()
```

```python
# Conteo por tipo de animal (Intakes)
plt.figure(figsize=(8,4))
sns.countplot(x="Animal Type", data=intakes,
              order=intakes["Animal Type"].value_counts().index)
plt.title("Distribuci√≥n por tipo de animal (Intakes)")
plt.show()

# Conteo por tipo de outcome
plt.figure(figsize=(8,4))
sns.countplot(x="Outcome Type", data=outcomes,
              order=outcomes["Outcome Type"].value_counts().index)
plt.title("Distribuci√≥n por tipo de outcome")
plt.xticks(rotation=45)
plt.show()
```

```python
# Convertir fechas si no lo hicimos antes
if "DateTime" in intakes.columns:
    intakes["DateTime"] = pd.to_datetime(intakes["DateTime"], errors="coerce")
    intakes["Year"] = intakes["DateTime"].dt.year

if "DateTime" in outcomes.columns:
    outcomes["DateTime"] = pd.to_datetime(outcomes["DateTime"], errors="coerce")
    outcomes["Year"] = outcomes["DateTime"].dt.year

# Agrupar por a√±o
intakes_year = intakes.groupby("Year").size().reset_index(name="Intakes Count")
outcomes_year = outcomes.groupby("Year").size().reset_index(name="Outcomes Count")

# Graficar
plt.figure(figsize=(10,5))
sns.lineplot(data=intakes_year, x="Year", y="Intakes Count", label="Ingresos (Intakes)")
sns.lineplot(data=outcomes_year, x="Year", y="Outcomes Count", label="Egresos (Outcomes)")
plt.title("üìà Evoluci√≥n de ingresos y outcomes por a√±o")
plt.xlabel("A√±o")
plt.ylabel("Cantidad de registros")
plt.legend()
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

def encode_dataframe(df):
    df_encoded = df.copy()
    for col in df_encoded.columns:
        if df_encoded[col].dtype == "object":
            df_encoded[col] = LabelEncoder().fit_transform(df_encoded[col].astype(str))
    return df_encoded

# Intakes
intakes_encoded = encode_dataframe(intakes)
plt.figure(figsize=(12,8))
sns.heatmap(intakes_encoded.corr(), cmap="coolwarm", center=0)
plt.title("üîó Matriz de correlaci√≥n - Intakes")
plt.show()

# Outcomes
outcomes_encoded = encode_dataframe(outcomes)
plt.figure(figsize=(12,8))
sns.heatmap(outcomes_encoded.corr(), cmap="coolwarm", center=0)
plt.title("üîó Matriz de correlaci√≥n - Outcomes")
plt.show()

# Licenses
licenses_encoded = encode_dataframe(licenses)
plt.figure(figsize=(12,8))
sns.heatmap(licenses_encoded.corr(), cmap="coolwarm", center=0)
plt.title("üîó Matriz de correlaci√≥n - Licenses")
plt.show()
```
